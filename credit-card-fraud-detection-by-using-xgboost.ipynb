# Credit Card Fraud Detection

## data from [kaggle](https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud/data)

信用卡公司能夠識別詐欺性信用卡交易非常重要，這樣客戶就不會因未購買的商品而被收取費用

特徵 V1、V2、…V28 是透過 PCA 獲得的主要成分，唯一未透過 PCA 轉換的特徵是「Time」和「Amount」

特徵「Time」包含資料集中每個事務與第一個事務之間經過的秒數

特徵「Amount」是交易金額，該特徵可用於範例相關的成本敏感學習

特徵「Class」是回應變量，如果有欺詐，則取值 1，否則取值 0

```python
# 計算每個 class 的資料數量
count_class_0 = df[df['Class'] == 0].shape[0]
count_class_1 = df[df['Class'] == 1].shape[0]

# 資料
sizes = [count_class_0, count_class_1]
labels = ['Class 0', 'Class 1']

# 圓餅圖
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Class')
plt.axis('equal')  # 使圓餅圖比例相等
plt.show()
```

```python
# 設定圖表字型大小
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=18)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)

# 移除 'Time' 和 'Class' 欄位，因為它們的分布通常不會被視為常態分佈
columns_to_plot = df.columns.drop(['Time', 'Class'])

# 計算子圖網格大小
num_cols = 4
num_rows = (len(columns_to_plot) - 1) // num_cols + 1

# 創建子圖
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 4*num_rows))

# 繪製每個欄位的分布圖
for i, column in enumerate(columns_to_plot):
    ax = axes[i // num_cols, i % num_cols]
    sns.histplot(df[column], kde=True, color='skyblue', ax=ax)
    ax.set_title(f'{column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')
    ax.grid(True)

# 移除多餘的子圖
for i in range(len(columns_to_plot), num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

plt.tight_layout()
plt.show()
```

```python
df = df.drop(['V13', 'V15', 'V22', 'V25', 'V26'], axis=1)
```

## 標準化

```python
from sklearn.preprocessing import StandardScaler

# 建立 StandardScaler 實例
std_scaler = StandardScaler()
```

```python
y = df['Class']
x = df.drop('Class', axis=1)
x = pd.DataFrame(std_scaler.fit_transform(x), columns=x.columns)
```

## Model

```python
from sklearn.model_selection import train_test_split

# stratify=y：確保訓練集和測試集都保持與原始資料集相同的詐欺案例比例。
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True , stratify=y, random_state=42)
```

```python
print("Training set Fraud:", sum(y_train == 1))
print("Testing set Fraud:", sum(y_test == 1))
```

```python
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score

xgb = XGBClassifier(random_state=42)

# 訓練模型
xgb.fit(x_train, y_train)

# 預測測試集
y_pred = xgb.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("\n")

# 產生分類報告
class_report = classification_report(y_test, y_pred)
print(class_report)
print('\n')

# 產生混淆矩陣
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from scipy.stats import uniform, randint

param_grid = {
    'n_estimators': list(range(5, 1000, 5)),
    'max_depth': list(range(1,20)),
    'random_state':[42]
}

rd_search = RandomizedSearchCV(xgb, param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)
rd_search.fit(x_train, y_train)

# 打印最佳超參數
print("Best parameters:", rd_search.best_params_)

# 使用最佳超參數訓練模型
xgb_best_model = XGBClassifier(**rd_search.best_params_)
xgb_best_model.fit(x_train, y_train)

# 預測測試集
xgb_y_pred = xgb_best_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, xgb_y_pred)
print("Accuracy:", accuracy)
print("\n")

# 產生分類報告
class_report = classification_report(y_test, xgb_y_pred)
print(class_report)
print('\n')

# 產生混淆矩陣
conf_matrix = confusion_matrix(y_test, xgb_y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
